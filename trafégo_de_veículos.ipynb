```python
from pycaret.datasets import get_data
get_data('index')
```

```python
dados = get_data('traffic')
```

```python

dados = dados.rename(columns={'Rush Hour': 'Rush_Hour'})
```

```python
dados.head(15)
```

```python
import seaborn as sns
sns.scatterplot(x = 'snow_1h', y = 'traffic_volume', hue = 'Rush_Hour', data = dados);
```

```python
from pycaret.regression import setup, models, create_model, tune_model, \
                               compare_models, blend_models, stack_models, \
                               predict_model, evaluate_model, \
                               finalize_model, save_model, load_model
```

```python
regressao = setup(data = dados, target = 'traffic_volume', normalize = True)
```

```python
models()
```

```python
lr = create_model('lr')
```

```python
rf_tunado = tune_model(lr, optimize = 'RMSLE',n_iter=10)
```

```python
melhores = compare_models(sort = 'RMSLE', n_select = 5)
```

```python
dt = create_model('dt')
```

```python
gbr_tunado = tune_model(dt, optimize = 'RMSLE', n_iter = 10)
```

```python
modelo_combinado = blend_models(melhores)
```

```python
modelo_stack = stack_models(melhores)
```

```python
pred = predict_model(gbr_tunado)
```

```python
round(pred,1)
```

```python
evaluate_model(rf_tunado)
```

```python
modelo_final = finalize_model(rf_tunado)
```

```python
save_model(modelo_final, 'meu-melhor-modelo-para-trafÃ©go-veiculos')
```

```python
import pandas as pd
modelo = load_model('meu-melhor-modelo-para-trafÃ©go-veiculos')

def Velocidade_trafego(Rush_Hour, temp, clouds_all, rain_1h, snow_1h,holiday,weather_main):
    dados0 = {'Rush_Hour': [Rush_Hour	], 'temp': [temp], 'clouds_all': [clouds_all], 'rain_1h': [rain_1h], 'snow_1h': [snow_1h],'holiday':[holiday],'weather_main':[weather_main]}
    dados = pd.DataFrame(dados0)

    pred = (float(predict_model(modelo, data = dados)['Label'].round(2))) 

    print('Velocidade do trafego = ',round(pred,2))
```
